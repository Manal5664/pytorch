```python
import torch
import time

# define te size of the matrices
size=10000 #large size for performance comparison

# create random matrices on CPU
mat_cpu1=torch.randn(size , size)
mat_cpu2=torch.randn(size , size)

# Measure time on cpu
start_time=time.time()
result_cpu=torch.matmul(mat_cpu1 , mat_cpu2)
cpu_time = time.time() - start_time

print(f"CPU time: {cpu_time:.4f} sec")

# Move matric to GPU
mat_gpu1=mat_cpu1.to('cuda')
mat_gpu2=mat_cpu2.to('cuda')


# Measure time on GPU

start_time=time.time()
result_gpu=torch.matmul(mat_gpu1 , mat_gpu2)
torch.cuda.synchronize()
gpu_time = time.time() - start_time

print(f"GPU time: {gpu_time:.4f} sec")

# compare

print("speedup (CPU time / GPU time): ", cpu_time/gpu_time)
```
